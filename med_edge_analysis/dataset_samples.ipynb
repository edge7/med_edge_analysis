# Dataset Samples Overview

3 sample questions from each dataset with full text.

```python
import json
from pathlib import Path
from IPython.display import display, Markdown

from constants import DATASETS, MODELS, MED_QA_DATASET

MED_QA_FOLDER = Path("/media/edge7/Extreme Pro/med_edge/benchmarks")
MED_AGENT = Path("/media/edge7/Extreme Pro/med_edge/benchmarks_test")
```

```python
def get_samples(dataset_name: str, n_samples: int = 3) -> list:
    """Load n_samples from a dataset."""
    model = MODELS[0]
    split = "test"
    
    if dataset_name == MED_QA_DATASET:
        filepath = MED_QA_FOLDER / f"{model}_{split}_raw.jsonl"
    else:
        filepath = MED_AGENT / f"{model}_medagents_{dataset_name}_{split}_raw.jsonl"
    
    samples = []
    if filepath.exists():
        with open(filepath, 'r', encoding='utf-8') as f:
            seen_ids = set()
            for line in f:
                if len(samples) >= n_samples:
                    break
                if line.strip():
                    record = json.loads(line)
                    if record['sample_id'] in seen_ids:
                        continue
                    seen_ids.add(record['sample_id'])
                    samples.append(record)
    return samples
```

```python
def format_options(options: dict) -> str:
    """Format options dict as readable text."""
    lines = []
    for key in sorted(options.keys()):
        lines.append(f"- **{key.upper()}**: {options[key]}")
    return "\n".join(lines)


def display_dataset_samples(dataset_name: str, samples: list):
    """Display samples with full text."""
    md = f"# {dataset_name}\n\n"
    
    if not samples:
        md += "*No data available*\n\n---\n"
        display(Markdown(md))
        return
    
    for i, sample in enumerate(samples, 1):
        question = sample.get('question', sample.get('prompt', 'N/A'))
        options = sample.get('options', {})
        ground_truth = sample.get('ground_truth', 'N/A')
        
        md += f"## Question {i}\n\n"
        md += f"{question}\n\n"
        
        if options:
            md += f"### Options\n\n{format_options(options)}\n\n"
        
        md += f"**Ground Truth: {ground_truth}**\n\n"
        md += "---\n\n"
    
    display(Markdown(md))
```

```python
for dataset_name in DATASETS:
    samples = get_samples(dataset_name, n_samples=3)
    display_dataset_samples(dataset_name, samples)
```

```python
import hashlib
from collections import defaultdict

def get_all_question_hashes():
    """Get hash of first 200 chars of each question, grouped by dataset."""
    hash_to_datasets = defaultdict(list)  # hash -> [(dataset, question_preview, ground_truth)]
    
    for dataset_name in DATASETS:
        model = MODELS[0]
        split = "test"
        
        if dataset_name == MED_QA_DATASET:
            filepath = MED_QA_FOLDER / f"{model}_{split}_raw.jsonl"
        else:
            filepath = MED_AGENT / f"{model}_medagents_{dataset_name}_{split}_raw.jsonl"
        
        if not filepath.exists():
            print(f"File not found: {filepath}")
            continue
        
        seen_ids = set()
        with open(filepath, 'r', encoding='utf-8') as f:
            for line in f:
                if line.strip():
                    record = json.loads(line)
                    if record['sample_id'] in seen_ids:
                        continue
                    seen_ids.add(record['sample_id'])
                    
                    question = record.get('question', record.get('prompt', ''))
                    ground_truth = record.get('ground_truth', 'N/A')
                    text_to_hash = question[:200]
                    q_hash = hashlib.md5(text_to_hash.encode()).hexdigest()
                    hash_to_datasets[q_hash].append((dataset_name, text_to_hash[:80], ground_truth))
    
    return hash_to_datasets

# Check for duplicates
hash_to_datasets = get_all_question_hashes()

duplicates = {h: entries for h, entries in hash_to_datasets.items() if len(entries) > 1}

# Filter to only cross-dataset duplicates
cross_dataset_duplicates = {}
for h, entries in duplicates.items():
    datasets_involved = set(e[0] for e in entries)
    if len(datasets_involved) > 1:
        cross_dataset_duplicates[h] = entries

if cross_dataset_duplicates:
    consistent = 0
    inconsistent = 0
    
    print(f"FOUND {len(cross_dataset_duplicates)} DUPLICATE QUESTIONS ACROSS DATASETS:\n")
    
    for h, entries in cross_dataset_duplicates.items():
        answers = set(e[2] for e in entries)
        is_consistent = len(answers) == 1
        
        if is_consistent:
            consistent += 1
            status = "✓ SAME ANSWER"
        else:
            inconsistent += 1
            status = "✗ DIFFERENT ANSWERS"
        
        print(f"Hash: {h} - {status}")
        for dataset, preview, gt in entries:
            print(f"  - {dataset}: answer={gt} | \"{preview}...\"")
        print()
    
    print("=" * 60)
    print(f"SUMMARY: {consistent} consistent, {inconsistent} INCONSISTENT")
    if inconsistent > 0:
        print("WARNING: Some duplicate questions have DIFFERENT ground truth answers!")
else:
    print("All questions are unique across datasets!")
```
